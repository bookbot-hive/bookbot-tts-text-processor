```python
from text_processor import TextProcessor
import os
import json

# All words in a single list
WORDS = [
    "eh", "oh", "erm", "ugh", "huh", "ahem", "ooh", "ahh", "mm-hmm", "uh-huh", "uh-oh", "hmm", "mmm",
    "sooo", "wellll", "ummm", "hmmm", "yesss", "alriiight", "okaaay", "riiight", "yeees", "yeaaah", 
    "ohhh", "uhhhh", "ehhhhh", "mmmmm", "hmmmmm", "oooookay", "exaaaactly", "oooh", "aaaah", "yessss", "errrr",
    "right", "okay", "yeah", "yes", "yep", "absolutely", "indeed", "certainly", "aha", "gotcha", 
    "fair enough", "you bet", "agreed", "understood", "all right", "roger", "affirmative", 
    "you betcha", "for sure", "well said", "yes indeed", "clearly", "no doubt", "got it", 
    "makes sense", "sure thing", "for sure", "yeah yeah", "okay okay",
    "Let me see", "well", "that's interesting", "that's an excellent point", "that's a great point"
]

model_dirs = {
        "en": "bookbot/roberta-base-emphasis-onnx-quantized",
        "sw": "",
        "id": ""
}
db_paths = {
    "en": "/home/s44504/3b01c699-3670-469b-801f-13880b9cac56/Emphasizer/data/words_emphasis_lookup_mixed.json",
    "sw": "",
    "id": ""
}
cosmos_config = {
    "url": os.getenv("COSMOS_DB_URL"),
    "key": os.getenv("COSMOS_DB_KEY"),
    "database_name": "Bookbot"
}

model = TextProcessor(model_dirs["en"], db_paths["en"], language="en", use_cosmos=True, cosmos_config=cosmos_config)
word2input_ids = {}
for word in WORDS:
    result = model.get_input_ids(word, phonemes=False, return_phonemes=True, push_oov_to_cosmos=True, add_blank_token=True)
    print(f"word: {word}, phonemes: {result['phonemes']}, input_ids: {result['input_ids']}")
    input_ids = result["input_ids"]
    word2input_ids[word] = input_ids

with open("word2input_ids.json", "w") as f:
    json.dump(word2input_ids, f)
```
